```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"font.size": 30})

cdc_data = pd.read_csv("NCHSData48.csv")
cdc_data

# annual population and annual death data
population_data = pd.read_excel("C:\\Users\\JLCat\\Learn-Python-for-Stats-and-Econ\\Projects\\COVID19\\U.S.PopulationAndAnnualDeathRate.xlsx", sheets = [0], index_col = [0])
population_data["Population"] = population_data["Population"].mul(10 ** 3)


# calculate weekly deats as pa portion of poulation
cdc_data["Population"] = np.NaN
for year in population_data.index:
    cdc_data["Population"][cdc_data["Year"] == year] = population_data.loc[year]["Population"]
    
cdc_data["Deaths / Population"] = cdc_data["All Deaths"].div(cdc_data["Population"])
cdc_data.dropna(inplace = True)


cdc_pivot= pd.pivot_table(cdc_data, 
               values=["All Deaths", "Deaths / Population"], 
               index=['Week'], columns=['Year'])
```

```python
fig, ax = plt.subplots(figsize = (40, 26))
cdc_pivot["Deaths / Population"].plot(ax = ax, legend = False)
ax.set_title("Deaths / Population\nWeekly")
plt.show()
plt.close()
```

```python
# replace 2020 death data with the sum from the weekly data. Multiply  
# by 12.5 / 11 to compensate for missing December and partial November data

population_data["Deaths per 100000"].loc[2020] = cdc_data[cdc_data["Year"]==2020]["All Deaths"].sum() / population_data.loc[2020]["Population"] * (12.5/11) * (10 ** 5) 
fig, ax = plt.subplots(figsize = (40, 26))
population_data["Deaths per 100000"].plot(ax =ax, legend = False, linewidth = 5)
last_year_above_2020 = population_data[population_data["Deaths per 100000"] > population_data["Deaths per 100000"].loc[2020]].index[-1]
ax.axvline(last_year_above_2020, color = "k", linestyle = "--")
ax.axhline(population_data.loc[2020]["Deaths per 100000"], color = "k", linestyle = "--")
ax.set_title("Deaths per 100000\nUnited States")
ax.set_ylim(ymin = 0)
ax.tick_params(axis='x', rotation=90)
```
